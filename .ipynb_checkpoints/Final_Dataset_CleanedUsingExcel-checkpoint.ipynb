```python
import pandas as pd
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt 
import re
import seaborn as sns

# Loading the data from the excel sheet

xls = pd.ExcelFile('CWQT_Master_DataSheet.xlsx')
df = pd.read_excel(xls, 'Cleaned_Data')

```

```python
df[["Month", "Year"]].loc[df[ "Year"] == 2018]["Month"].value_counts()
```

```python
# yearly = df["Year"].value_counts()

# yearly.to_csv('yearly.csv', index=True)
```

```python
print(len(df))
df = df.drop(df[(df["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"] == "<10") \
                | (df["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"] == ">24196") \
                | (df["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"] == ">6000")].index)

df = df.drop(["Notes"], axis=1)
df = df.drop(["Battery High Tide"], axis=1)
df = df.drop(["Sample Time"], axis=1)
df = df.dropna()
print(len(df))
```

```python
weekdays = []

for i in df.columns[5:13]:
    weekdays.append(i) 
    
df["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"] = df["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"].astype(int)
for i in weekdays:

    df[i] = df[i].astype(float)
```

```python
grouped_yearly_mean = df.groupby("Year")[['Most Probable Number (MPN) of Enterococcus colonies per 100 ml','Day of Collection Precipitation (Thursday)', 'Previous Day Precipitation (Wednesday)','Previous Tuesday Precipitation','Previous Monday Precipitation','Previous Sunday Precipitation','Previous Saturday Precipitation','Previous Friday Precipitation']].agg('mean')
 

grouped_yearly_mean  
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

sns.lineplot(data = grouped_yearly_mean , x=grouped_yearly_mean.index, y='Most Probable Number (MPN) of Enterococcus colonies per 100 ml', marker='o', ax=ax1)
ax2 = ax1.twinx()
```

```python
high_MPN = df.loc[df['Most Probable Number (MPN) of Enterococcus colonies per 100 ml'] > 104]
high_MPN_mean =  high_MPN.groupby("Year")[['Most Probable Number (MPN) of Enterococcus colonies per 100 ml','Day of Collection Precipitation (Thursday)', 'Previous Day Precipitation (Wednesday)','Previous Tuesday Precipitation','Previous Monday Precipitation','Previous Sunday Precipitation','Previous Saturday Precipitation','Previous Friday Precipitation']].agg('mean')
high_MPN_mean
```

```python
high_MPN_mean =  high_MPN.groupby("Year")[['Most Probable Number (MPN) of Enterococcus colonies per 100 ml','Day of Collection Precipitation (Thursday)', 'Previous Day Precipitation (Wednesday)','Previous Tuesday Precipitation','Previous Monday Precipitation','Previous Sunday Precipitation','Previous Saturday Precipitation','Previous Friday Precipitation']].agg('mean')
high_MPN_mean  
```

```python
grouped_yearly_mean["mean"] = grouped_yearly_mean[weekdays].mean(axis=1)
```

```python
grouped_yearly_mean.to_csv('mean_prep_yearly.csv', index=True)
```

```python
mean_prep = grouped_yearly_mean[weekdays].mean(axis=1)
mean_prep_df = pd.DataFrame(mean_prep, columns=["Prep"])
```

```python
high_MPN_mean["Regular Average"] = mean_prep_df['Prep']

high_MPN_mean['High Average'] = high_MPN_mean[weekdays].mean(axis=1)

high_MPN_mean.to_csv('regvshighMPN.csv', index= True)

```

```python
 def conditions(s):
    if (s["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"]  > 104):
        return "High"
    elif ((s["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"]  < 104) & (s["Most Probable Number (MPN) of Enterococcus colonies per 100 ml"]  >34) ):
        return "Unacceptable if levels persist"
    else:
        return "Acceptable"
```

```python
    
df['Status'] = df.apply(conditions, axis=1)
```

```python
df[["Site", "Status", "Most Probable Number (MPN) of Enterococcus colonies per 100 ml", "Year"]].to_csv("status.csv", index=False)
```
